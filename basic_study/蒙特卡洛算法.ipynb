```python
M = input('')  #
N=0                              #
 
import math
import random
 
for i in range(int(M)):
    x = random.random()          #random
    y = random.random()
    if math.sqrt(x**2+y**2)<1:   #
        N = N+1                  #
 
pi=4*N/int(M)
 
print(pi)
 
```

```python
from scipy import integrate

# 
def f(x):
    return x**2

# 
integral, error = integrate.quad(f, 0, 1)

print("", integral)
```
